#### 컬럼 설명

데이터셋은 데이터 과학 직무의 급여에 관련된 11개의 열로 구성되어 있습니다.


- work_year: 급여가 지급된 연도입니다.
- experience_level: 해당 연도 동안의 직무 경력 수준입니다.
- employment_type: 직무의 고용 형태입니다.
- job_title: 해당 연도 동안 근무한 직무(직책)입니다.
- salary: 총 지급된 총 급여액입니다.
- salary_currency: 급여가 지급된 통화를 ISO 4217 통화 코드로 나타냅니다.
- salaryinusd: 급여를 미국 달러로 환산한 금액입니다.
- employee_residence: 직원의 주요 거주 국가(ISO 3166 국가 코드)입니다.
- remote_ratio: 원격으로 수행한 작업의 전체 비율입니다.
- company_location: 고용주의 본사 국가 또는 계약 지점의 국가입니다.
- company_size: 해당 연도 동안 회사에서 일한 사람들의 중간 인원수입니다.

````python
# 직무 개수 및 직무컬럼의 데이터 구하기
print('직무개수 : ', len(df['job_title'].unique()))
df['job_title'].unique()
````

````python
# 직무별 임금 평균 구하기
average_salary = df.groupby('job_title')['salary_in_usd'].mean()
average_salary
````

````python
# 가장 높은 연봉 구하기
max_salary = average_salary.max()
print(max_salary)

# 가장 높은 연봉의 직무는 무엇일까?
max_salary_job = average_salary.idxmax()
print(max_salary_job)
````

````python
# 직무별 상위 10% 연봉 계산
top_salaries = df.sort_values('salary_in_usd', ascending=False)
result = top_salaries.groupby('job_title').head(1).head(10)
result
````

````python
job_titles = result['job_title'].tolist()
experience = result['experience_level'].tolist()
salarys = result['salary_in_usd'].tolist()

for job, exp, sal in zip(job_titles, experience, salarys):
    print(f'직무 : {job}, 경력 : {exp}, 연봉 : {sal}')
````
